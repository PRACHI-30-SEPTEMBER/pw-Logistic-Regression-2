### Q1. What is the purpose of grid search CV in machine learning, and how does it work?

Grid Search CV is used to find the best hyperparameters for a model.
It works by trying all possible combinations of parameters and evaluating them using cross-validation.

### Q2. Describe the difference between grid search CV and randomized search CV, and when might you choose one over the other?

- Grid Search: checks all parameter combinations → more accurate but slower
- Randomized Search: checks random combinations → faster but may miss best

Use Randomized when dataset is large or parameter space is big.

### Q3. What is data leakage, and why is it a problem in machine learning? Provide an example.

Data leakage occurs when information from the test set is used during training.
Example: using future data in training → model performs unrealistically well.

### Q4. How can you prevent data leakage when building a machine learning model?

- Proper train-test split
- Use pipelines
- Avoid using future data
- Apply preprocessing only on training data

### Q5. What is a confusion matrix, and what does it tell you about the performance of a classification model?

A confusion matrix is a table showing actual vs predicted values:
- True Positive (TP)
- True Negative (TN)
- False Positive (FP)
- False Negative (FN)

It helps evaluate classification performance.

### Q6. Explain the difference between precision and recall in the context of a confusion matrix.

- Precision = TP / (TP + FP) → how many predicted positives are correct
- Recall = TP / (TP + FN) → how many actual positives are detected

### Q7. How can you interpret a confusion matrix to determine which types of errors your model is making?

- High FP → model gives many false alarms
- High FN → model misses actual cases
- Helps identify model weaknesses

### Q8. What are some common metrics that can be derived from a confusion matrix, and how are they calculated?

- Accuracy = (TP + TN) / Total
- Precision = TP / (TP + FP)
- Recall = TP / (TP + FN)
- F1-score = 2 * (Precision * Recall) / (Precision + Recall)

### Q9. What is the relationship between the accuracy of a model and the values in its confusion matrix?

Accuracy depends on TP and TN compared to total predictions.
High accuracy may be misleading in imbalanced datasets.

### Q10. How can you use a confusion matrix to identify potential biases or limitations in your machine learning model?

By checking error distribution:
- If one class has more FN → bias against that class
- Helps identify imbalance and fairness issues
